```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('paises_datos_integrados.csv')
```

```python
# Encontrar los 5 países más baratos
paises_baratos = df.sort_values('costo_promedio_total').head(5)

# Encontrar los 5 países más caros
paises_caros = df.sort_values('costo_promedio_total', ascending=False).head(5)

print("Los 5 destinos más económicos para viajar:")
display(paises_baratos[['pais', 'continente', 'costo_promedio_total']])

print("\nLos 5 destinos más costosos para viajar:")
display(paises_caros[['pais', 'continente', 'costo_promedio_total']])

plt.figure(figsize=(12, 8))

# Combinar los datos
comparacion = pd.concat([paises_baratos, paises_caros])
# Marcar categoría (económico/costoso)
comparacion['categoria'] = ['Económico']*5 + ['Costoso']*5

# Crear gráfico
ax = sns.barplot(x='costo_promedio_total', y='pais', hue='categoria', 
                data=comparacion, palette=['#2ecc71', '#e74c3c'])

plt.title('Contraste entre los 5 destinos más económicos y más costosos', fontsize=16)
plt.xlabel('Costo promedio total (USD)', fontsize=12)
plt.ylabel('País', fontsize=12)

for i, v in enumerate(comparacion['costo_promedio_total']):
    ax.text(v + 3, i, f"${v:.0f}", va='center')

plt.tight_layout()
plt.show()
```

El análisis revela un contraste marcado entre los destinos más económicos y los más costosos. Los países más accesibles (Namibia, Grecia, Dinamarca, Moldova y Tanzania) tienen un costo promedio de entre $67 y $93 USD, mientras que los más costosos (Bolivia, Chile, Bangladesh, Panamá y Canadá) oscilan entre $219 y $247 USD.

La diferencia entre Namibia ($67) y Bolivia ($247) representa una brecha de casi 3.7 veces en los costos, lo que sugiere oportunidades significativas de ahorro dependiendo de la elección del destino.

Es interesante observar que Europa, generalmente percibida como costosa, cuenta con tres países entre los cinco más económicos (Grecia, Dinamarca y Moldova), mientras que América domina entre los destinos más caros con cuatro países (Bolivia, Chile, Panamá y Canadá). Esto contradice la percepción común de que viajar a Europa es siempre más costoso que viajar a otros destinos como Sudamérica.

```python
# Calcular promedios globales de cada categoría de gasto
hospedaje_promedio = df['hospedaje_precio_promedio_usd'].mean()
comida_promedio = df['comida_precio_promedio_usd'].mean()
transporte_promedio = df['transporte_precio_promedio_usd'].mean()
entretenimiento_promedio = df['entretenimiento_precio_promedio_usd'].mean()

# Calcular porcentajes del gasto total
costo_total = hospedaje_promedio + comida_promedio + transporte_promedio + entretenimiento_promedio
porcentaje_hospedaje = (hospedaje_promedio / costo_total) * 100
porcentaje_comida = (comida_promedio / costo_total) * 100
porcentaje_transporte = (transporte_promedio / costo_total) * 100
porcentaje_entretenimiento = (entretenimiento_promedio / costo_total) * 100

# Mostrar los resultados
distribucion = pd.DataFrame({
    'Categoría': ['Hospedaje', 'Comida', 'Transporte', 'Entretenimiento'],
    'Costo Promedio (USD)': [hospedaje_promedio, comida_promedio, transporte_promedio, entretenimiento_promedio],
    'Porcentaje del Total': [porcentaje_hospedaje, porcentaje_comida, porcentaje_transporte, porcentaje_entretenimiento]
})

distribucion['Porcentaje del Total'] = distribucion['Porcentaje del Total'].round(1)
distribucion['Costo Promedio (USD)'] = distribucion['Costo Promedio (USD)'].round(2)

display(distribucion)

# Crear visualización
plt.figure(figsize=(10, 8))
categorias = ['Hospedaje', 'Comida', 'Transporte', 'Entretenimiento']
valores = [hospedaje_promedio, comida_promedio, transporte_promedio, entretenimiento_promedio]

# Crear gráfico
plt.pie(valores, labels=categorias, autopct='%1.1f%%', startangle=90, 
    shadow=False, explode=(0.05, 0, 0, 0), 
    colors=['#3498db', '#f1c40f', '#2ecc71', '#9b59b6'])

# Añadir título
plt.title('Distribución del presupuesto de viaje promedio mundial', fontsize=16)
plt.axis('equal')
plt.tight_layout()
plt.show()

```

La distribución del presupuesto de viaje muestra que el hospedaje constituye el mayor componente del gasto total, representando un 32.6% ($45.83 USD) del presupuesto promedio. La comida ocupa el segundo lugar con un 26.2% ($36.87 USD), seguida por el entretenimiento con 21.9% ($30.79 USD) y el transporte con 19.4% ($27.23 USD).

Esta distribución sugiere que las estrategias de ahorro centradas en el hospedaje tendrían el mayor impacto en el presupuesto total de viaje. La diferencia relativamente equilibrada entre las demás categorías (comida, entretenimiento y transporte) indica que los viajeros deben considerar todos estos aspectos al planificar sus gastos, pero priorizar la búsqueda de alojamientos económicos para maximizar su ahorro.

Para optimizar el presupuesto de viaje, los viajeros podrían considerar alternativas de hospedaje más económicas como hostales, alojamientos compartidos o ubicaciones fuera de las zonas turísticas principales, lo que podría reducir significativamente el costo total de su viaje.



```python
# Calcular costo promedio por continente
costos_continente = df.groupby('continente')['costo_promedio_total'].mean().sort_values(ascending=False)

# Mostrar resultados
display(pd.DataFrame({'Continente': costos_continente.index, 'Costo Promedio (USD)': costos_continente.values.round(2)}))

# Identificar el continente más económico y el más costoso
continente_caro = costos_continente.index[0]
continente_barato = costos_continente.index[-1]
diferencia_porcentual = ((costos_continente.iloc[0] - costos_continente.iloc[-1]) / costos_continente.iloc[-1]) * 100

print(f"Viajar a {continente_caro} es en promedio {diferencia_porcentual:.1f}% más costoso que viajar a {continente_barato}")

# Crear visualización
plt.figure(figsize=(12, 7))
ax = sns.barplot(x=costos_continente.index, y=costos_continente.values, palette='viridis')
plt.title('Costo promedio de viaje por continente', fontsize=16)
plt.xlabel('Continente', fontsize=12)
plt.ylabel('Costo promedio total (USD)', fontsize=12)
plt.xticks(rotation=0)

# Añadir los valores en las barras
for i, v in enumerate(costos_continente):
    ax.text(i, v + 5, f"${v:.0f}", ha='center')

plt.tight_layout()
plt.show()
```

Los costos promedio por continente revelan diferencias significativas en los presupuestos de viaje necesarios según la región del mundo. América se posiciona como el continente más costoso con un promedio de $180 USD, seguido por Asia ($156 USD), Europa ($128 USD) y África ($115 USD), que resulta ser el más económico.

La diferencia entre América y África es considerable: viajar por América resulta aproximadamente 56.5% más costoso que hacerlo por África. Esta variación permite a los viajeros ajustar sus expectativas presupuestarias según la región que deseen visitar.

Este hallazgo contradice la percepción común de que Europa es el continente más caro para viajar, ubicándose en realidad en tercer lugar. Los altos costos en América podrían explicarse por la presencia de destinos premium como Canadá, Estados Unidos y varios países sudamericanos que aparecen entre los más costosos del data set, como Bolivia y Chile.

Para viajeros con presupuesto limitado, África ofrece la mejor relación costo-beneficio a nivel continental, mientras que quienes deseen visitar América deberían planificar un presupuesto significativamente mayor o buscar destinos específicos más económicos dentro del continente.
